```python
#display charts using seaborn library in python
```

```python
#loading titanic dataset
import seaborn as sns
titanic = sns.load_dataset('titanic')
print("Dimensions of the dataset is: ", titanic.shape)
print("The columns  of the dataset are:\n", titanic.keys())
print("The details of the dataset are:\n", titanic.describe())
print("The info of the dataset are:\n",titanic.info())
```

```python
#plots for categorical data
#1.box plot
sns.boxplot(x='class',y='fare',data=titanic,order=["Second","Third","First"],palette="Set1").set_title('Box plot for fare with respect to class')
```

```python
#2.violin plot
sns.violinplot(x='class',y='age',hue='sex',data= titanic).set_title('Violin plot for age with respect to sex')
```

```python
#use violin plot to display different categories of embarked with age and use pallete option also.
sns.violinplot(x='embarked',y='age',data= titanic,palette=['red','purple','yellow']).set_title('Violin plot for age with respect to embarked')
```

```python
#3.point plot
sns.pointplot(x='class',y='age',data= titanic).set_title('Point plot for age and class')
```

```python
#4.line plot
sns.lineplot(x='embarked',y='fare',data= titanic).set_title('Line plot for fare and embarked')
```

```python
#5.count plot
sns.countplot(y='alive',data= titanic,order=['yes','no'],palette=['orange','green'],orient='h').set_title('Count plot for alive')
```

```python
#6.bar plot
sns.barplot(x='fare',y='embarked',data= titanic,hue= 'embarked').set_title('Bar plot for fare with respect to embarked')
```

```python
#7.strip plot
sns.stripplot(x="class" , y="fare",jitter=True,data = titanic,palette="Set1",hue = "class",legend=False).set_title("Strip plot for class and fare")
```

```python
#8.swarm plot
sns.swarmplot(x="class" , y="fare",data = titanic,palette="Set1",hue = "class",legend=False).set_title("Swarm plot for class and fare")
```

```python
#create a single chart displaying the bar plot and swarm plot together but of different colors to understand the importance of both charts effectively
```

```python
import matplotlib.pyplot as plt
plt.figure(1,figsize=(10,5))
plt.subplot(1,2,1)
sns.barplot(x="class",y="age",data= titanic,palette="Set1").set_title("Bar plot for class and age")
plt.subplot(1,2,2)
sns.swarmplot(x="class",y="age",data= titanic,palette="Set1").set_title("Swarm plot for class and age")

plt.show()
```

```python
#9.cat plot
#cat plot with default settings including hue
sns.catplot(x='sex',y='age',hue='alive',data= titanic)
#cat plot with kind ="violin"
sns.catplot(x='pclass',y='fare',hue='alive',kind = "violin",data= titanic)
#cat plot with kind ="box" and "class"
sns.catplot(x='class',y='age',hue='alive',kind = "box",data= titanic)
#cat plot with kind ="bar",col="class" and col_wrap=3
sns.catplot(x='class',y='fare',hue='alive',col="deck",col_wrap=4,kind ="bar",data= titanic)
```

```python
#10.Facet grid
#creating a facet grid
sns.FacetGrid(titanic,col="class")
#creating a facet grid of histogram for age & class
grid1 = sns.FacetGrid(titanic,col="class")
grid1.map(plt.hist,"age")
#creating a facet grid of histogram for age,class and alive
grid1 = sns.FacetGrid(titanic,col="class",hue="alive")
grid1.map(plt.hist,"age")
#creating a facet grid of scatter chart for age,sex and fare
grid2 = sns.FacetGrid(titanic,col="class",hue="sex")
grid2.map(plt.scatter,"age","fare")
#creating a facet grid of scatter chart for age,sex and fare
grid3 = sns.FacetGrid(titanic,col="deck",hue="sex")
grid3.map(plt.hist,"age")
```
