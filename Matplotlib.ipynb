```python
#to display multiple lines with different shapes and colors
```

```python
a = list(range(1,11))
b = list(range(5,55,5))
c = list(range(10,110,10))
d = list(range(20,210,20))
print("First list: " , a)
print("Second list: " , b)
print("Third list: " , c)
print("Fourth list: " , d)
plt.plot(a,a,'g^',a,b,'bs',a,c,'r--',a,d,'mo')
plt.xlim(0,11)
plt.ylim(0,220)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Display multiple lines with different shapes and colors')
plt.grid(True,color = 'k')
plt.show()
```

```python
#creating a pie chart using matplotlib in python
```

```python
#pie chart is created using pie()
list1 = [10,30,50,14,15,40,16,17,56,80,90,88]
plt.pie(list1,labels = list1)
plt.show()
```

```python
#to display a pie chart with labels inside
items = 'Pizza','Burger','Pasta','Chowmin','Momos','Finger chips','Spring Roll','Samosa'
consumers = [20,15,30,40,80,10,70,90]

fig, ax = plt.subplots()
ax.pie(consumers, labels=items,autopct='%1.1f%%')
plt.show()
```

```python
#to display patterns instead of labels inside
items = 'Pizza','Burger','Pasta','Chowmin','Momos','Finger chips','Spring Roll','Samosa'
consumers = [20,15,30,40,80,10,70,90]

fig, ax = plt.subplots()
ax.pie(consumers, labels=items, hatch=['**','oo','++','OO','\\','//','XX','..'])
plt.show()
```

```python
import numpy as np
p = [10,14,16,27,88,79,100,14,56,100,90]
q = [12,18,20,30,85,74,95,12,58,45,94]
r = [15,19,22,35,89,75,90,15,59,48,97]
s = [18,17,25,38,84,74,100,14,57,50,92]
print('List p: ',p)
print('List q: ',q)
print('List r: ',r)
print('List s: ',s)

data = [p,q,r,s]
colors = ['r','b','m','g']
plt.violinplot(data,showmedians=True)
plt.grid(True,color='k')
plt.title("Different violin plots on the same chart")
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
ecommerce = ['Myntra', 'Snapdeal', 'Meesho', 'Amazon', 'Flipkart']
Q1_profit = [35,45,100,75,50]
Q2_profit = [10,80,56,87,70]
Q3_profit = [95,25,10,70,40]
Q4_profit = [85,5,70,60,90]

plt.scatter(ecommerce,Q1_profit,color = 'm')
plt.scatter(ecommerce,Q2_profit,color = 'b')
plt.scatter(ecommerce,Q3_profit,color = 'r')
plt.scatter(ecommerce,Q4_profit,color = 'g')
plt.title('Scatter plot of ecomerce companies profit')
plt.xlabel('Organizations')
plt.ylabel('Profit')
plt.show()
```

```python
#creating bar chart using matplotlib in python
ecommerce = ['Myntra', 'Snapdeal', 'Meesho', 'Amazon', 'Flipkart']
Q1_profit = [35,45,100,75,50]
Q2_profit = [10,80,56,87,70]
Q3_profit = [95,25,10,70,40]
Q4_profit = [85,5,70,60,90]

plt.figure(1, figsize = (10,10))
plt.subplot(221)
plt.bar(ecommerce,Q1_profit)
plt.title('Quater 1 profit')
#creating bar chart in 2nd cell
plt.subplot(222)
plt.bar(ecommerce,Q2_profit)
plt.title('Quater 2 profit')
#creating bar chart in 3rd cell
plt.subplot(223)
plt.bar(ecommerce,Q3_profit)
plt.title('Quater 3 profit')
#creating bar chart in 4rth cell
plt.subplot(224)
plt.bar(ecommerce,Q4_profit)
plt.title('Quater 4 profit')
plt.show()
```

```python
import numpy as np
a = [10,20,30,40,50,60,70]
b = [60,40,80,100,120,140,50]
plt.figure(1, figsize=(15,15))
#creating quiver plot in 1st of image having 3 rows and 2 columns
plt.subplot(3,2,1)
plt.quiver(a,b,color='red')
#creating quiver plot in 2nd cell
plt.subplot(3,2,2)
plt.quiver(b,a,color='blue')
#creating quiver plot in 3rd cell
plt.subplot(3,2,3)
x= 8
y= 6
M,N = np.mgrid[0:x,0:y]
plt.quiver(M,N,color='cyan')
#creating quiver in 4th cell
plt.subplot(3,2,4)
plt.quiver(M,N,color='violet')
#creating quiver in 5th cell
plt.subplot(3,2,5)
x= 20
y= 15
M,N = np.mgrid[0:x,0:y]
plt.quiver(M,N,color='orange')
#creating quiver in 6th cell
plt.subplot(3,2,6)
plt.quiver(N,M,color='brown')
plt.suptitle('Quiver plots')
plt.show()
```

```python
#creating mesh grid using matplotlib in python
#meshgrid()
```

```python
#creating large datasets
val1 = np.linspace(-9,9,600)
val2 = np.linspace(-5,5,500)
#creating a meshgrid chart
A,B = np.meshgrid(val1,val2)
plt.figure(1,figsize = (10,5))
plt.subplot(1,3,1)
plt.imshow(A+B)
plt.subplot(1,3,2)
plt.imshow(A-B)
plt.subplot(1,3,3)
plt.imshow(A*B)

plt.suptitle('Mesh grid')
```

```python
#creating contour plot using matplorlib in python
#contour()
```

```python
import numpy as np
#creating large datasets
val1 = np.linspace(-9,9,600)
val2 = np.linspace(-5,5,500)
#creating a mesh grid
A,B = np.meshgrid(val1,val2)
plt.figure(1,figsize=(15,10))
plt.subplot(2,4,1)
plt.contour(A,B,A+B,alpha = .75,cmap = 'viridis') #cmap can be:viridis,viridis,jet,cividis
plt.subplot(2,4,2)
plt.contourf(A,B,A+B,alpha = .75,cmap = 'viridis')
plt.subplot(2,4,3)
plt.contour(A,B,A-B,alpha = .75,cmap = 'viridis')
plt.subplot(2,4,4)
plt.contourf(A,B,A-B,alpha = .75,cmap = 'viridis')
plt.subplot(2,4,5)
plt.contour(A,B,A*B,alpha = .75,cmap = 'viridis')
plt.subplot(2,4,6)
plt.contourf(A,B,A*B,alpha = .75,cmap = 'viridis')
plt.subplot(2,4,7)
plt.contour(A,B,B%A,alpha = .75,cmap = 'viridis')
plt.subplot(2,4,8)
plt.contourf(A,B,B%A,alpha = .75,cmap = 'viridis')

plt.suptitle('Contour plots')
plt.show()
```
