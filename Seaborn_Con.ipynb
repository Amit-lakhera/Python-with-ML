```python
#seaborn library for visualizing continous variables
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
print("Dataset is:\n",iris.head())
print("Number of dimensions in the dataset are:",iris.shape)
```

```python
#scatter plot
sns.scatterplot(y="sepal_length",x="petal_length",data=iris,hue="species").set_title('Scatter plot for sepal_length and petal_length')
```

```python
#regression plot
#a)regplot()
sns.regplot(y='petal_length',x='petal_width',data=iris)
#b)lmplot
sns.lmplot(y='sepal_length',x='petal_width',data=iris)
```

```python
#using lmplot for higher order "sepal_width","petal_width"
sns.lmplot(y='sepal_width',x='petal_width',data=iris,order=2)
```

```python
#heat map
sns.heatmap(iris[['petal_length','petal_width']]).set_title('Heat map for petal_length and petal_width')
```

```python
#with specified colors
sns.heatmap(iris[['petal_length','petal_width']],cmap="RdYlBu").set_title('Heat map for petal_length and petal_width')
```

```python
#4.univariate distribution plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

#univariate distribution plot for petal_length
plt.subplot(311)
sns.distplot(iris[['petal_length']]).set_title('Distribution plot for petal_length')

#histogram for "petal_length"
plt.subplot(312)
sns.distplot(iris[['petal_length']],kde=False).set_title('Distribution plot for kde=false')

#kernel density estimation plot for "petal_length"
plt.subplot(313)
sns.distplot(iris[['petal_length']],hist=False).set_title('Distribution plot for hist=false')
```

```python
#5.creating displot chart using seaborn library
sns.displot(iris[['sepal_length','sepal_width','petal_length','petal_width']],palette='Set1')
```

```python
#6.joint plot -bivariate analysis
#joint plot for sepal_length and petal_length
sns.jointplot(y='petal_length',x='sepal_length',data=iris,palette='Set1')
```

```python
#regression plot using joint plot
sns.jointplot(y='petal_length',x='sepal_length',kind='reg',data=iris)
```

```python
sns.jointplot(y='petal_width',x='sepal_width',kind='hist',data=iris)
```

```python
#7.joint hexbin plot for bivriate data analysis
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='hex',color="red") #dark color-lower density #light color-higher density
```

```python
#8.joint kernel density plot for bivariate data analysis
sns.jointplot(x='sepal_length',y='petal_length',data=iris,kind='kde',color="green")
```

```python
#9.pairplot for multivariate data analysis
#pair plot using regression line-selected cols of dataset
sns.pairplot(iris,vars=['sepal_length','sepal_width','petal_length','petal_width'],hue='species',kind='reg')
#pair plot for different species-selected cols of dataset
sns.pairplot(iris,vars=['sepal_length','sepal_width','petal_length','petal_width'],hue='species',kind='hist',palette='colorblind')
```

```python
#pair plot using regression line for the entire dataset
sns.pairplot(iris,kind='reg')
#pair plot for different species and kde as diagonal plot
sns.pairplot(iris,hue='species',kind='scatter',diag_kind='kde',palette='deep')

```

```python
#10.pair grid for multivariate data analysis
#creating pair grid with scatter plot
grid1 = sns.PairGrid(iris,hue='species')
grid1.map(plt.scatter)
#creating pair grid with scatter,kde and hist plot
grid2 = sns.PairGrid(iris,hue='species')
grid2.map_upper(plt.scatter)
grid2.map_lower(sns.kdeplot)
grid2.map_diag(plt.hist)
```
