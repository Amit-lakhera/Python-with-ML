```python
#Logistic Regression using Binary Classification
#to find out the person at the age of 65 and 24 want to buy insurance or not
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'/content/insurance_data.csv')
```

```python
dataset
```

```python
dataset['bought_insurance'].replace({'no':'0','yes':'1'} , inplace = True)
```

```python
dataset
```

```python
#visualizing the data points
plt.scatter(x='age' , y = 'bought_insurance' , data = dataset)
```

```python
#model training
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(dataset[['age']],dataset['bought_insurance'] , test_size=0.2)
```

```python
#importing the logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```
